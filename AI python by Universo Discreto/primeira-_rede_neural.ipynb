```python
import numpy as np
import matplotlib.pyplot as plt 

np.random.seed(42) #it will define the random numbers, so when the firs nuymber generate it will stay into the same random numbers, it´s like a seed of minecraft that makes you in the same world map default in the seed 
ages = np.random.randint(low=15, high=70,size= 40)

print(ages)
```

```python
label = []

for age in ages:
    if age < 30:
        label.append(0)
    else:
        label.append(1)


for i in range(0,4):
    numberRandom = np.random.randint(0, len(label))
    if label[numberRandom] == 1:
        label[numberRandom] = 0
    else:
        label[numberRandom] = 1
```

```python
plt.scatter(ages, label, color="red")
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(ages.reshape(-1, 1), label)
```

```python
m = model.coef_[0]
b = model.intercept_

print(m,b)
```

# REGRESSÃO LINEAR DAQUELE CONJUNTO DE PONTOS 

```python
linear_idade = (0.5 - b) / m
print(linear_idade)

plt.plot(ages, ages * m + b, color="blue")
plt.scatter(ages.reshape(-1,1), label)
plt.show()
```
