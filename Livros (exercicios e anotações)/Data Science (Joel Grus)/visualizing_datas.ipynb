```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gpd = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a grafio of lines, years is  in axle x and GPD is in axle y  
plt.plot(years, gpd, color='green', marker='o', linestyle='solid')

plt.title("Nominla GPD")
plt.ylabel("Billions of $")
plt.xlabel("years")

plt.show()
```

```python
series = ["LCDP", "BM", "WUAP", "TF", "LP"]
times_user_watched_show = [3,9,1,4,7]

# create a grafic bar with cordenates x for left [0,1,2,3,4], height [3,9,1,4,7] (the height of each bar)
plt.bar(range(len(series)), times_user_watched_show)

plt.title("My Favorite Seris")
plt.ylabel("Most watched shows (user)")

# put the labels in the center
plt.xticks(range(len(series)),series)

plt.show(),
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
population = [7.05, 7.12, 7.19, 7.27, 7.35, 7.43, 7.51, 7.60, 7.70, 7.79]

plt.plot(years, population, color="black", marker ="o", linestyle="solid")

plt.xlabel("Years")
plt.ylabel("Population by billions")

plt.show()
```

```python
years = [15, 22, 17, 30, 19, 25, 16, 28, 21, 18]
more_age = [idade >= 18 for idade in years] # testa todas as idades cada uma recebera ou True ou False

plt.bar(range(len(years)), more_age) # create a bar with the x being the years and the y being if they are adults or not (years >= 18)

plt.xlabel("Years")
plt.ylabel("More Age")

plt.xticks(range(len(years)), years) # put the values in the center 

plt.show()
```
