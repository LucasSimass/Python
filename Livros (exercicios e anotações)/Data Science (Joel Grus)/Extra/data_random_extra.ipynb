```python
from collections import Counter
from random import randint,seed
from matplotlib import pyplot as plt

seed(10) # random.randint

random_numbers = Counter(randint(1,1000) for i in range(1000000))


print(random_numbers)

plt.bar(
    x=random_numbers.keys(),
    height=random_numbers.values(),
    width=6,
    edgecolor=(0,0,0,0)
)


plt.axis([-50, 1050, 0, max(random_numbers.values()) + 1000])

plt.show()
```

```python
from collections import Counter
from random import randint,seed
from matplotlib import pyplot as plt

seed(10) # random.randint

random_numbers: list[tuple[int]] = [(randint(1,1000) for i in range(2)) for i in range(100000)]

result: Counter[list[int]] = []
for a,b in random_numbers: 
    result += [a + b] 

result = Counter(result)

plt.bar(
    x=result.keys(),
    height=result.values(),
    color="blue",
    width=3,
)

plt.ylabel("Count")
plt.xlabel("Sum of two Random Numbers")

plt.axis([-100, max(result.keys()) + 100, 0, max(result.values()) + 10])

plt.show()


```
