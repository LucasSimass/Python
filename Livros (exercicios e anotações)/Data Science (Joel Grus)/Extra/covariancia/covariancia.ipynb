```python
import matplotlib.pyplot as plt 

x = [4,6,11,16,3]
y = [50,50,40,30,60]

plt.scatter(x=x,y=y, marker="o", c="red")
plt.axis([min(x) - 1, max(x) + 1, min(y) - 20, max(y) + 5])
plt.yticks([0,10,20,30,40,50,60])

plt.show()

# calcula a media
def media(dados: list[float]) -> float:
    assert dados, "A lista não pode estar vazia"
    return sum(i for i in dados) / len(dados)

# o quanto cada elemento esta desviado da media
def desvio(dados: list[float]) -> list[float]:
    assert dados, "A lista não pode estar vazia"
    return [i - media(dados) for i in dados]

# covariancia representa uma ligação entre x e y e analisa uma possivel relação entre eles 
# o valor da covariancia não indica magnitude, só se importe se o valor tiver negativo ou positivo
# se covariancia > 0 = Yi tende a aumentar toda vez que Xi aumenta
# se covariancia < 0 = Yi tende a diminuir toda vez que Xi aumenta
def covariancia(xs: list[float], ys: list[float]) -> float:
    assert xs, "A lista xs não pode estar vazia"
    assert ys, "A lista ys não pode estar vazia"
    assert len(xs) == len(ys), "As listas devem conter a mesma quantidade de numeros"
    return sum((x - media(xs)) * (y - media(ys)) for x, y in zip(xs,ys)) / (len(xs) - 1)

print(desvio(x))
print(desvio(y))
print(covariancia(xs=x, ys=y))
```
