```python
# the bars grafic is usefull to do data science about quantity of something

from matplotlib import pyplot as plt
series = ["LCDP", "BM", "WUAP", "TF", "LP"]
times_user_watched_show = [3,9,1,4,7]

# create a grafic bar with cordenates x for left [0,1,2,3,4], height [3,9,1,4,7] (the height of each bar)
plt.bar(range(len(series)), times_user_watched_show)

plt.title("My Favorite Series")
plt.ylabel("Most watched shows (user)")

# get the labels and put it in the center
plt.xticks(range(len(series)), series)

plt.show(),
```

```python
from matplotlib import pyplot as plt
years = [15, 22, 17, 30, 19, 25, 16, 28, 21, 18]
older_age = [idade >= 18 for idade in years]     # testa todas as idades cada uma recebera ou True ou False

plt.bar(range(len(years)), older_age)            # create a bar with the x being the years and the y being if they are adults or not (years >= 18)

plt.xlabel("Years")
plt.ylabel("More Age")

plt.xticks(range(len(years)), years)                # put the values in the center 

plt.show()
```

```python
from matplotlib import pyplot as plt
from collections import Counter

grades: list[int] = [83, 72, 96, 0, 100, 54, 78, 81, 96, 67, 57, 43,50]

                                            # agroupe the grades by decile, but put the 100 with the 90
                                            #the // return only the integer (if has a float it descart)
histogram: Counter[int]= Counter(min(grade // 10 * 10, 90) for grade in grades) # it will return a map

plt.bar(
    [x + 5 for x in histogram.keys()],      # move the bars to the right in five
    histogram.values(),                     # atribute the for each bar the correct height
    10,                                     # atribute the width ten for each bar
    edgecolor=(0,0,0),                      # turn dark the border of the bars
    )

plt.axis([-5,105,0,5])                      # axle of x betwenn -5 and 105
                                            # axle of y betwenn 0 and 5

plt.xticks([10 * i for i in range(11)],)    # labels

plt.xlabel("Grades")
plt.ylabel("Decile")

plt.title("Distribution of Exam 1 Grades")

print(histogram)                            # Counter({90: 3, 50: 3, 80: 2, 70: 2, 0: 1, 60: 1, 40: 1})
print(min(0,90))                            # return the minimun value possible in the list 
print(100 // 10 * 10)
plt.show()
```

```python
from matplotlib import pyplot as plt
mentions: list[int] = [500,505]
years: list[int] = [2017, 2018]

plt.bar(
    x=years,
    height=mentions,
    width=0.8,
    edgecolor=(0,0,0)
)

plt.xticks(years)

plt.ylabel("Num of times that i head \"Data Science\" in the year")
plt.xlabel("Years")

# if you dont do this, matplotlib will label the axle x with 0 , 1 
# and adding a +2.013e3 off in the corner 
plt.ticklabel_format(useOffset=False)

plt.axis([2016.5, 2018.5, mentions[0] - 1, mentions[1] + 1 ])
print([*mentions, *years])
plt.show()
```
