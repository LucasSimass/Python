```python
from matplotlib import pyplot as plt
from collections import Counter

grades: list[int] = [83, 72, 96, 0, 100, 54, 78, 81, 96, 67, 57, 43,50]

                                            # agroupe the grades by decile, but put the 100 with the 90
                                            #the // return only the integer (if has a float it descart)
histogram: Counter[int]= Counter(min(grade // 10 * 10, 90) for grade in grades) # it will return a map

plt.bar(
    [x + 5 for x in histogram.keys()],      # move the bars to the right in five
    histogram.values(),                     # atribute the for each bar the correct height
    10,                                     # atribute the width ten for each bar
    edgecolor=(0,0,0),                      # turn dark the border of the bars
    )

plt.axis([-5,105,0,5])                      # axle of x betwenn -5 and 105
                                            # axle of y betwenn 0 and 5

plt.xticks([10 * i for i in range(11)],)    # labels

plt.xlabel("Grades")
plt.ylabel("Decile")

plt.title("Distribution of Exam 1 Grades")

print(histogram)                            # Counter({90: 3, 50: 3, 80: 2, 70: 2, 0: 1, 60: 1, 40: 1})
print(min(0,90))                            # return the minimun value possible in the list 
print(100 // 10 * 10)
plt.show()
```

```python
import matplotlib.pyplot as plt
from random import randint,seed
from collections import Counter

seed(10) # it will always show the same random data

num_friends = [randint(0,100) for i in range(101)]
num_friends.append(100)

friends_counts = Counter(num_friends)

xs = range(101) # Number of friends
ys = [friends_counts[x] for x in xs]

plt.bar(xs, ys)

plt.axis([0, 101, 0, 5])

plt.title("Histrogram of Friends Counts")

plt.xlabel("Number of connections")
plt.ylabel("Number of people")

plt.show()

print(num_friends)
print(Counter(num_friends))
print(ys)

num_point = len(num_friends)

largest_value = max(num_friends)
smallest_value = min(num_friends)

sorted_values = sorted(num_friends)

smallest_value = sorted_values[0]
second_smallest_value = sorted_values[1]
second_largest_value = sorted_values[-1]

# retorna o valor médio da lista
def mean(xs: list[float]) -> float:
    return sum(xs) / len(xs)

print(mean(xs=num_friends))

# the underline it means that the function is private so the users cant acess
# only can be called by the median function, but not for other users
def _median_odd(xs: list[float]) -> float:
    """Se len(xs) for impar, a mediana será o elemento do meio"""
    return sorted(xs)[len(xs) // 2]

print(_median_odd(xs=num_friends))

def _median_even(xs: list[float]) -> float:
    sorted_xs = sorted(xs)
    hi_midpoint = len(xs) // 2
    return (sorted_xs[hi_midpoint - 1] + sorted_xs[hi_midpoint]) / 2

print(_median_even(num_friends))

def median(v: list[float]) -> float:
    """Encontre o valor do meio v"""
    return _median_even(v) if len(v) % 2 == 0 else _median_odd(v) 

assert median([1,10,2,9,5]) == 5, "the function return should be 5"
assert median([3,4,2,1,5,0]) == 2.5, "the function returun should be 2.5"

print(median(num_friends))

def quantile(xs: list[float], porcentage: float) -> float:
    """Retorna o valor pht-percentile em x"""
    p_index = int(porcentage * len(xs))
    return sorted(xs)[p_index]

assert quantile([1,2,3,4,5], 0.1) == 1
assert quantile([1,2,3,4,5], 0.25) == 2
assert quantile([1,2,3,4,5], 0.50) == 3
assert quantile([1,2,3,4,5], 0.75) == 4
assert quantile([1,2,3,4,5], 0.90) == 5

# numeros que mais apareceram na lista de amigos
def mode(xs: list[float]) ->list[float]:
    """Retorna uma lista, pois pode gaver mais de uma moda"""
    counts = Counter(xs)
    max_count = max(counts.values())
    return [x_i for x_i,counts in counts.items() if counts == max_count if counts == max_count]

print(Counter(num_friends)) 
print(mode(num_friends)) # [4, 30]

def data_range(xs: list[float]) -> float:
    return max(xs) - min(xs)

assert data_range([1,2,4,5,100]) == 99

# todos os valores são subtraidos pelo valor medio da lista, criando uma visão bem mais clara e simplifica de onde está o valor médio e seus valores; exemplo: [1,2,3,4,5] = [-2,-1,0,1,2]
def de_mean(xs: list[float]) -> list[float]:
    """Traduza xs subtraindo sua média (para que o resultado tenha média)"""
    x_bar = mean(xs)
    return [ x - x_bar for x in xs]

print(de_mean([1,2,3,4,5,6,7,8,9])) # [-4.0, -3.0, -2.0, -1.0, 0.0, 1.0, 2.0, 3.0, 4.0]

def variance(xs: list[float]) -> float:
    """Quase o desvio quadrado médio da média"""
    assert len(xs) >= 2, "variance requires at least two elemenets"

    n = len(xs)
    devitation = de_mean(xs)
    return sum_of_squares(devitation) / (n - 1)

# eleva os números ao quadrado e depois soma eles
def sum_of_squares(xs: list[float]) -> float:
    return sum(n * n for n in xs)

print(sum_of_squares([5,10])) # 5^2 + 10^2 = 125 
assert sum_of_squares([5,2]) == 29, "5^2 + 5^2 = 50" 

print(variance(num_friends)) # 726.4346534653466

from math import sqrt

def standart_devitation(xs: list[float]) -> float:
    """O desvio padrão é a raiz quadrada da variância"""
    return sqrt(variance(xs))

print(standart_devitation(xs)) # 29.300170647967224

def interquartile_range(xs: list[float]) -> float:
    """Retorna a diferença entre o percentil 75% e percentil 25%"""
    return quantile(xs, 0.75) - quantile(xs, 0.25)

print(interquartile_range(num_friends)) # 44

def dot(xs: list[float], ys: list[float]) -> float:

    """it will find the dot in the dot product"""

    assert len(xs) == len(ys), "Should have the same size"
    return sum(x * y  for x,y in zip(xs, ys))

def covariance(xs: list[float], ys: list[float]) -> float:
    assert len(xs) == len(ys), "xs and ys must have the same lenght"

    return dot(de_mean(xs), de_mean(ys)) / (len(xs) - 1)

print(covariance(xs=num_friends, ys=num_friends))

def correlation(xs: list[float], ys: list[float]) -> float:
    """Mede a variação simultanea de xs e ys a partir das suas médias"""
    stdev_x = standart_devitation(xs)
    stdev_y = standart_devitation(ys)
    if stdev_x > 0 and stdev_y > 0:
        return covariance(xs,ys) / stdev_x / stdev_y
    else:
        return 0
    
print(correlation(xs,ys)) # -0.09894610641341027

friends: list[int] = [75, 65, 72, 63, 71, 64, 60, 64, 70,]
minutes: list[int] = [175, 170, 205, 120, 220, 130, 105, 145, 190]

outlier = num_friends.index(100)

print(outlier) # 101

num_friends_good = [x
                    for i, x in enumerate(num_friends)
                    if i != outlier]

print(sorted(num_friends_good)) # [0, 1, 4, 4, 4, 4, 5, 5, 8..9]

daily_minutes_good = [x
                      for i, x in enumerate(minutes)
                      if i != outlier]
```
