```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gpd = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a grafio of lines, years is  in axle x and GPD is in axle y  
plt.plot(years, gpd, color='green', marker='o', linestyle='solid')

plt.title("Nominal GPD")
plt.ylabel("Billions of $")
plt.xlabel("years")

plt.show()
```

```python
from matplotlib import pyplot as plt
variance = [2 ** i for i in range(9)] # [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256 / (2 ** i) for i in range(9)] # you dont need the (), but we gonne use to be more ez to read
total_errors = [x + y for x, y in zip(variance, bias_squared)] # [257.0, 130.0, 68.0, 40.0, 32.0, 40.0, 68.0, 130.0, 257.0]
xs = [index for index, values in enumerate(variance)] # [0, 1, 2, 3, 4, 5, 6, 7, 8]

print(total_errors) # enumerate gives us the index and value to use in the for

# we can multiple calls for plt.plot
# to show multiples series in th same grafic
plt.plot(xs, variance, "g-", label="variance") # "g-" == green solid line  (color and line style)
plt.plot(xs, bias_squared, "r-", label="bias^2") # "r-" == red solid line (color and line style)

plt.plot(xs, total_errors, "b:", label="total error") # "b:" == blue dotted line (color and line style)

# like we atributes labes each serie
# podemos cirar uma legenda de graa (loc=9 means "top center")
plt.legend(loc=9) # loc=9 == top center
plt.xlabel("model complexity")
plt.xticks([])
plt.title("The Bias-Variance TradeOff")
plt.show()
```
